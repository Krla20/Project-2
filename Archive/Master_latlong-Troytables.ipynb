### Extraction

```python
# Store csv into df
csv_file= "../Resources/alt_school_rating_list_tea.csv"
raw_school_rating_df= pd.read_csv(csv_file)
raw_school_rating_df.head()
```

### Transformation

```python
# removing columns not needed
raw_school_rating1_df = raw_school_rating_df.drop(['ESC\r\nRegion', 'Region', 'SBOE\r\nDistrict','Campus 2020 Rating',
                                                  'Campus 2018 Rating', 'Campus 2017 Rating','Campus 2016 Rating',
                                                  'Campus 2015 Rating', 'Campus 2014 Rating','Campus 2013 Rating',
                                                  'Campus 2012 Rating','Campus 2011 Rating'], axis =1)
raw_school_rating1_df
```

```python
# Renaming columns - int is for intermediate (in process) table
int_school_rating_df= raw_school_rating1_df.rename(columns= {'Campus\r\nNumber': 'campus_number',
                                                            'Campus' : 'school',
                                                            'District' : 'district',
                                                          '# of Consecutive Years Campus is Academically Unacceptable' : 'yrs_unacceptable',
                                                          'Campus 2019 Rating' : 'rating'})
int_school_rating_df
```

```python
# Adding in new column and associated value
int_school_rating_df['year']= 2019
int_school_rating_df
```

```python
# Dropping rows that contain notes, i.e. start with *
int_school_ratinga_df = int_school_rating_df.drop(int_school_rating_df.loc[int_school_rating_df.campus_number.str.startswith("*")].index)
int_school_ratinga_df
```

```python
# Dropping rows that contain middle school to match HAR data
int_school_rating1_df = int_school_ratinga_df.drop(int_school_ratinga_df.loc[int_school_ratinga_df.school.str.contains("MIDDLE")].index)
int_school_rating1_df
```

```python
# checking for NANs
int_school_rating1_df.isna().sum()
```

```python
# filling NaNs
int_school_rating1_df = int_school_rating1_df.fillna("Not Rated")
```

```python
# looking for unique items in column
int_school_rating1_df.rating.unique()
```

## create tables for grouped zip and area scatter charts

```python
tyPlayraw_df = HAR1_df

# create new df where year_built = 0 are dropped
tyPlay_df= tyPlayraw_df[tyPlayraw_df['year_built'] != 0]

# allows all columns to display
pd.set_option('display.max_columns', None)

tyPlay_df
```

```python
# goal of these steps are to created a df that is grouped on zipcode
# that can be used in scatter plot
zipPriceGroup_df = tyPlay_df.groupby('zip').list_price.agg(['count','min', 'max', 'mean', 'std'])
zipPriceGroup_df['mean'] = zipPriceGroup_df['mean'].astype(int)
zipPriceGroup_df.rename(columns = {'count': 'count_zip','min': 'min_price', 'max': 'max_price', 
                                   'mean': 'avg_price', 'std': 'std_price'}, inplace = True)
zPG_reset_df = zipPriceGroup_df.reset_index()
zPG_reset_df
```

```python
zPG_fix = zPG_reset_df[zPG_reset_df['std_price'].notna()]
zPG_fix['std_price'] = zPG_fix['std_price'].astype(int)
zPG_fix.head()
```

```python
zipAgeGroup_df = tyPlay_df.groupby('zip').year_built.agg(['mean', 'std'])
zipAgeGroup_df['mean'] = zipAgeGroup_df['mean'].astype(int)
zipAgeGroup_df.rename(columns = {'mean': 'avg_age', 'std' : 'std_age'}, inplace = True)
zAG_reset_df = zipAgeGroup_df.reset_index()
zAG_reset_df
```

```python
zAG_fix = zAG_reset_df[zAG_reset_df['std_age'].notna()]
zAG_fix['std_age'] = zAG_fix['std_age'].astype(int)
zAG_fix.head()
```

```python
zipFinalraw_df = pd.merge(zPG_fix, zAG_fix, on="zip")
zipFinal_df = zipFinalraw_df.sort_values(by='count_zip', ascending = False)
zipFinal_df.head(20)
```

```python
# goal of these steps similiar to zip aboce, but instead
# of grouping on zipcode, this will create a table grouped by area
areaPriceGroup_df = tyPlay_df.groupby('area').list_price.agg(['count','min', 'max', 'mean', 'std'])
areaPriceGroup_df['mean'] = areaPriceGroup_df['mean'].astype(int)
areaPriceGroup_df.rename(columns = {'count' : 'count_area', 'min': 'min_price', 'max': 'max_price', 
                                    'mean': 'avg_price', 'std' : 'std_price'}, inplace = True)
aPG_reset_df = areaPriceGroup_df.reset_index()
# aPG_reset_df.to_csv("../Resources/aPG_reset_df.csv", index = False)

aPG_reset_df.head()
```

```python
# import table that has area (neighborhood)  and associated abbbreviation.
# abbreviation will be used inside circles in scatter plot

csv_abbrtab = "../Resources/abbr_table.csv" 
abbr_table_df= pd.read_csv(csv_abbrtab)
# abbr_table_df
```

```python
# merge df for abbreviations 
aInt = pd.merge(aPG_reset_df, abbr_table_df, on="area")
# aInt
```

```python
aPG_fix = aInt[aInt['std_price'].notna()]
aPG_fix['std_price'] = aPG_fix['std_price'].astype(int)
aPG_fix.head()
```

```python
areaAgeGroup_df = tyPlay_df.groupby('area').year_built.agg(['mean', 'std'])
areaAgeGroup_df['mean'] = areaAgeGroup_df['mean'].astype(int)
areaAgeGroup_df.rename(columns = {'mean': 'avg_age', 'std' : 'std_age'}, inplace = True)
aAG_reset_df = areaAgeGroup_df.reset_index()
aAG_reset_df.head()
```

```python
aAG_fix = aAG_reset_df[aAG_reset_df['std_age'].notna()]
aAG_fix['std_age'] = aAG_fix['std_age'].astype(int)
aAG_fix.head()
```
